# Seaborn - Statistical data visualization tool

## 1.What is Seaborn?
Seaborn is a python library that builts on the top of the matplotlib that makes the statistical data more easy and beautiful.

### Why we need it?
1.For creating plots like violin plot,box plot,pair plots

2.Automatically handles data from pandas dataframes

### Where we use it?
1.Data analysis and reporting

2.Exploring relationships between variables

3.Machine learning for visualizing datasets and results

## 2. Importing Seaborn

import seaborn as sns

```python
import seaborn as sns
```

```python

import matplotlib.pyplot as plt
import pandas as pd
 
```

## Histogram

```python
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True,color='gray')
plt.title("Histogram of Age")
plt.show()
 

 
```

## Count plot

```python
# Countplot
sns.countplot(x='class', data=df,color='gray')
plt.title("Count of Passengers by Class")
plt.show()
```

## Scatter plot

```python
### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
 
```

## Boxplot

```python

# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
```

## Violin plot

```python
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
 
```

## Pair plot

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
 
```

## Heat map

```python
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

## Customization & Themes

```python

sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='gray',edgecolor='black')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

## Bar plot

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

## KDE plot of fare distribution

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='gray')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```

#### 1.Compare Passenger Survival by Class and Gender

```python
survival_rates = df.groupby(['class', 'sex'])['survived'].mean().reset_index()
sns.barplot(x='class', y='survived', data=survival_rates, hue='sex', ci=None)
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Class and Sex')
plt.show()
```

#### 2. Distribution of Passenger Ages

```python
sns.kdeplot(df['age'].dropna(), fill=True, color='gray')
plt.title("KDE Plot of age Distribution")
plt.show()
```

#### 3. Fare Distribution by Class

```python

# Boxplot
sns.boxplot(x='class', y='fare', data=df)
plt.title("Boxplot of fare by Class")
plt.show()
 
```

#### 4. Relationship Between Age and Fare

```python
sns.scatterplot(data=df, x='age', y='fare',hue='class')
plt.show()
```

####  5. Correlation Heatmap


```python
corr = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
